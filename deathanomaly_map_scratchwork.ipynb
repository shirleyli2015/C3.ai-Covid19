```
import plotly.graph_objects as go
import numpy as np
import pandas as pd
pd.set_option('display.max_columns', 50 )
pd.set_option('display.max_rows', 50)
```

```
deaths = pd.read_csv('/content/drive/My Drive/Colab Notebooks/Weekly_counts_of_death_by_jurisdiction_and_cause_of_death.csv')
weighteddeath = deaths.loc[deaths['Type'] == "Predicted (weighted)"]
wd_uptoweek = weighteddeath.loc[weighteddeath['Week']<=41]
sum_wd = wd_uptoweek.groupby(["Time Period","Year","Cause Group","State Abbreviation"]).sum()
dd=sum_wd.groupby(["Time Period","Cause Group","State Abbreviation"]).mean()
dd= dd['Number of Deaths']
old = dd[(dd.index.get_level_values(0) == '2015-2019') & (dd.index.get_level_values(1) == "Alzheimer disease and dementia")]
new = dd[(dd.index.get_level_values(0) == '2020') & (dd.index.get_level_values(1) == "Alzheimer disease and dementia")]
old_noak=old.drop(labels =['US','AK'], level=2)
new_noak=new.drop(labels =['US','AK'], level=2)
#2020 lacked alaska data!
#seems like some states have incomplete reporting!!!!
d_anom = new_noak.subtract(old_noak.array)
fig = go.Figure(data=go.Choropleth(
    locations=d_anom.index.get_level_values(2), # Spatial coordinates
    z = d_anom.array, # Data to be color-coded
    locationmode = 'USA-states', # set of locations match entries in `locations`
    colorscale = 'Purples',
    colorbar_title = "# of people",
))

fig.update_layout(
    title_text = 'Excess deaths of Alzheimer disease and dementia in 2020 compared to 2015-2019 Averages up to Week 41',
    geo_scope='usa', 
)

fig.show()
```
